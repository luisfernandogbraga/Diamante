```python
# vamos ver se temos alguns outliers (dados dispersos)
with sns.axes_style('whitegrid'):
    grafico= sns.boxplot(x=diamond['price'])
    grafico.set(title='PreÃ§os OUTLIIERS do Diamante', xlabel='PreÃ§o')
```

**Insight do grÃ¡fico **: No caso desse grÃ¡fico, podemos ver que existem muitos valores outliers acima da caixa, o que indica que existem alguns diamantes com preÃ§os bem acima da mÃ©dia. Esses outliers podem ser interessantes de serem investigados para entender melhor o que os diferencia dos outros diamantes. A mÃ©dia de preÃ§o desses diamantes estÃ¡ perto da casa dos 2500 UDS

CorrelaÃ§Ã£o entre o do preÃ§o do diamante com seu peso agrupados por sua transparÃªcia.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamond, x='carat', y='price', hue='clarity')
plt.title('PreÃ§o da Diamente em funÃ§Ã£o do peso agrupados por transparÃªncia')
plt.xlabel('Peso')
plt.ylabel('PreÃ§o')
```

Podemos observar que existem vÃ¡rios pontos concentrados na parte inferior do grÃ¡fico, indicando que hÃ¡ muitos diamantes com baixo peso e preÃ§o mais baixo. Por outro lado, hÃ¡ alguns diamantes com alto peso e alto preÃ§o, que aparecem como pontos isolados na parte superior direita do grÃ¡fico.
