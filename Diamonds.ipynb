```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamond= sns.load_dataset('diamonds')
display(diamond)
```

```python
diamond.info()
```

```python
#bucando dados nulos
nulos= diamond.isnull().sum()
nulos
```

```python
# vamos ver se temos alguns outliers (dados dispersos)
with sns.axes_style('whitegrid'):
    grafico= sns.boxplot(x=diamond['price'])
    grafico.set(title='Preços OUTLIIERS do Diamante', xlabel='Preço')
```

**Insight do gráfico **: No caso desse gráfico, podemos ver que existem muitos valores outliers acima da caixa, o que indica que existem alguns diamantes com preços bem acima da média. Esses outliers podem ser interessantes de serem investigados para entender melhor o que os diferencia dos outros diamantes. A média de preço desses diamantes está perto da casa dos 2500 UDS

Correlação entre o do preço do diamante com seu peso agrupados por sua transparêcia.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamond, x='carat', y='price', hue='clarity')
plt.title('Preço da Diamente em função do peso agrupados por transparência')
plt.xlabel('Peso')
plt.ylabel('Preço')
```

Podemos observar que existem vários pontos concentrados na parte inferior do gráfico, indicando que há muitos diamantes com baixo peso e preço mais baixo. Por outro lado, há alguns diamantes com alto peso e alto preço, que aparecem como pontos isolados na parte superior direita do gráfico.
